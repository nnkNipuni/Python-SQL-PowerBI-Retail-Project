### Handling Missing Review Ratings
Some products have missing review ratings, which can distort analysis if left untreated. 
Instead of removing these records or using a single global value, missing ratings are filled using the median rating within each product category.

For example, if an Electronics product has a missing rating, it is replaced with the median rating of all Electronics products rather than the overall dataset median. 
This preserves category-specific rating behavior while reducing the impact of outliers.


```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
                                                                        
```

### Creating Age Groups
Categorize customers by age using 4 quantiles with labels:

```python
#Create a column  as age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

### Purchase Frequency Feature Engineering
- Creating Purchase Frequency in Days -
Convert the `frequency_of_purchases` column into the **number of days** between purchases using a predefined mapping.  
This creates a numeric column `purchase_frequency_days` for easier analysis of purchase intervals.


```python
# Creating the column -   purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
